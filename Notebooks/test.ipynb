```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/5/raw.csv', header=None,
                 names=["timestamp", "VOLT", "PW", "PWI", "PPB", "FREQ", "calibration_counter"])

df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')
print(df['datetime'])
```

```python
plt.figure(figsize=(12, 8))

for col in ["VOLT", "PW", "PWI", "PPB", "FREQ"]:
    plt.plot(df['datetime'], df[col], label=col)

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Input Values Over Time")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
columns_to_plot = ["VOLT", "PW", "PWI", "PPB", "FREQ"]

for col in columns_to_plot:
    plt.figure(figsize=(10, 4))
    plt.plot(df['datetime'], df[col])
    plt.title(f"{col} over time")
    plt.xlabel("Time")
    plt.ylabel(col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
