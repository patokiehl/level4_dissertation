```python
from garmin_fit_sdk import Decoder, Stream
import matplotlib.pyplot as plt
import pandas as pd
```

```python
stream = Stream.from_file("FIT_files/participant_HR_5.fit")
decoder = Decoder(stream)
messages, errors = decoder.read()
```

```python
heart_rate_data = []
for record in messages.get("record_mesgs", []):
    if "heart_rate" in record:
        heart_rate_data.append({
            "timestamp": record["timestamp"],
            "heart_rate": record["heart_rate"]
        })

first_50 = heart_rate_data
heart_rates = [entry["heart_rate"] for entry in first_50]

start_time = first_50[0]["timestamp"]
print(start_time)
elapsed_time = [(entry["timestamp"] - start_time).total_seconds() for entry in first_50]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(elapsed_time, heart_rates, linestyle='-', color="red")
plt.xlabel("Elapsed Time (s)")
plt.ylabel("Heart Rate")
plt.title("First 50 Heart Rate Measurements")
plt.ylim(0, 150) 
plt.tight_layout()
plt.show()
```
